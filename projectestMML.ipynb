```python
import pandas as pd

# Load the selected dataset
input_csv_file = r"C:\Users\1\Downloads\MLL\RFEMLL_50_percent_features.csv"  # Replace with your actual file path
selected_df = pd.read_csv(input_csv_file)

# Separate the features and target variable
X_selected = selected_df.drop(columns=['Class'])
y_selected = selected_df['Class']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y_selected, test_size=0.2, random_state=42)

from sklearn.neighbors import KNeighborsClassifier  # Remove the extra quotation mark

# Create a kNN classifier with the desired value of 'k'
k = 3  # You can choose an appropriate value for 'k'
knn_classifier = KNeighborsClassifier(n_neighbors=k)
```

```python
from sklearn.model_selection import cross_val_score

Knn_scores = []

for K in range(1, 21):
    knn_classifier = KNeighborsClassifier(n_neighbors=K)
    score = cross_val_score(knn_classifier, X, y, cv=10)
    Knn_scores.append(score.mean())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming you have Knn_scores defined
# For demonstration, let's generate random scores for Knn_scores
Knn_scores = np.random.rand(20)  # Replace with your actual scores

# Sort Knn_scores and get the corresponding indices
sorted_indices = np.argsort(Knn_scores)[::-1]
sorted_Knn_scores = Knn_scores[sorted_indices]

plt.figure(figsize=(15, 8))  # Adjust the figure size as needed

plt.plot([K for K in range(1, 21)], sorted_Knn_scores, color='red')
for i in range(1, 21):
    plt.text(i, sorted_Knn_scores[i - 1], f'{sorted_Knn_scores[i - 1]:.2f}', ha='center', va='bottom', fontsize=12)  # Adjust the fontsize as needed

plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier Scores for Different k Levels (Ranked)')
plt.grid(True)  # Add grid lines for better visualization

plt.show()

```
